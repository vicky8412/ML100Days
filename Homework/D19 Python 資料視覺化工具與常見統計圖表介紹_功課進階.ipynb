```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('F:/boston.csv')
data = pd.DataFrame(data)
data
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?

data.boxplot(figsize = (15, 4))
plt.xticks(rotation = 'vertical')
plt.ylim(0, 800)

median = data.median().between(300, 400)
print(median)
```

```python
TAX以及B欄位的中位數位於300~400之間
```

```python
#畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
x = data['NOX']
y = data['DIS']
plt.scatter(x, y, color = 'b', alpha = 0.3, marker = 'o')
plt.xlabel('NOX')
plt.ylabel('DIS')
plt.title('NOX vs DIS')
plt.show()
```

```python
兩者成反比關係
```
