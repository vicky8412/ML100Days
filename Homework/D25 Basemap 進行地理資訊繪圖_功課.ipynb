```python
# 導入開發套件
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

# 新建地圖
map = Basemap(projection='merc',llcrnrlon=-180, urcrnrlon=180, llcrnrlat=-80, urcrnrlat=80)
# 經度 : longitude 、 緯度 : latitude
# llcrnrlon – 左下角的經度。
# urcrnrlon – 右上角的經度。
# llcrnrlat – 左下角的緯度。
# urcrnrlat – 右上角的緯度。


# 畫圖
map.drawcoastlines()

# 顯示结果
plt.show()

# 存储结果
plt.savefig('test.png')
```

```python
map = Basemap(projection = 'ortho', lat_0 = 0, lon_0 = 0)
#’ortho’指正射投影，後面两個参數是设置中心點

# 给整個地圖上藍色
map.drawmapboundary(fill_color = 'aqua')

# 给陸地塗上珊瑚色，湖泊塗上藍色
map.fillcontinents(color = 'coral', lake_color = 'aqua')

# 畫圖
map.drawcoastlines()

# 顯示结果
plt.show()
```

```python
# 創建基本地圖。
# 當 "projection" 參數為“ortho"時，所得圖位地球儀截面

# 設定圖形大小
plt.figure(figsize=(8, 6))

# 設定投影型態
m=Basemap(projection = 'ortho', lat_0 = 60, lon_0 = -100)

# 設定圖形顏色, 選用NASA 寶石藍輸出 
m.bluemarble()

# 秀圖
plt.show()
```

```python
# 設置Lambert保形底圖。
# width=12000000,height=9000000
# 經緯度lat_1=45.,lat_2=55,lat_0=50,lon_0=-107
# set resolution = None跳過邊界數據集的處理
# projection='lcc'

m=Basemap(projection = 'lcc',
          resolution = None,
          width = 12000000, height = 9000000,
          lat_1 = 45, lat_2 = 55,
          lat_0 = 50, lon_0 = -107)

# 選用NASA 寶石藍輸出 
m.bluemarble()

# 秀圖
plt.show()
```

```python
#  "projection" 參數為“Mill"
# 設定經緯座標
# 左下角的緯度 - llcrnrlat = -90,
# 左下角的經度 - llcrnrlon = -180,
# 右上角的緯度 - urcrnrlat = 90,
# 右上角的經度 -urcrnrlon = 180
# 成圖設定resolution

m = Basemap(projection = 'mill',
          resolution = 'i',
          llcrnrlat = -90, llcrnrlon = -180,
          urcrnrlat = 90, urcrnrlon = 180,
          lat_0 = 50, lon_0 = -107)

#創建邊線
m.drawcoastlines()

#畫出國家，並使用線寬為2 的線條生成分界線。
m.drawcountries(linewidth = 2)

#畫出州界
m.drawstates(color = 'b')

#畫出城市
m.drawcounties(color = 'darkred')

plt.title('Basemap Tutorial')
plt.show()
```
